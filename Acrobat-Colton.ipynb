## Particle Size Stats

```python
def fivenum(v, label):
    """Returns Tukey's five number summary (minimum, lower-hinge, median, upper-hinge, maximum) for the input vector, a list or array of numbers based on 1.5 times the interquartile distance"""
    import numpy as np
    from scipy.stats import scoreatpercentile
    try:
        np.sum(v)
    except TypeError:
        print('Error: you must provide a list or array of only numbers')
    q1 = scoreatpercentile(v,25)
    q3 = scoreatpercentile(v,75)
    iqd = q3-q1
    md = np.median(v)
    whisker = 1.5*iqd
   
    return print('Five Number Summary for:', label,'\n','Min:', np.min(v),'Q1:', md-whisker, 'Median:', md, 'Q3', md+whisker, 'Max', np.max(v),'\n')
```

```python
medians = []
for i in plot:
    fivenum(acrobat[i],i)
    medians.append(np.median(acrobat[i]))
```

```python
plt.plot(medians)
plt.title('Particle Medians')
```

```python
acrobat['Particle Mean'] = acrobat.mean(axis=1)
acrobat['Particle Median'] = acrobat.median(axis=1)
acrobat['Mean Biomass'] = acrobat.mean(axis=1)
acrobat.plot(y='Particle Mean', use_index=True)
acrobat.plot(y='Particle Median', use_index=True)
acrobat.plot(y='Mean Biomass', use_index=True)
```

## Binning

```python
acrobat_binned = merged.groupby(merged.index // 10).sum()
acrobat_binned['Seconds from Start'] = acrobat_binned.index*5+5
acrobat_binned.set_index('Seconds from Start').head()
```

see percentages/stats of how the different sizes are affected by the parameters

```python
acrobat_binned["Total Bio"] = (acrobat_binned['Tiny'] + acrobat_binned['Small'] + acrobat_binned['Medium'] + acrobat_binned['Large'])
acrobat_binned['Percent Tiny'] = acrobat_binned['Tiny'] / acrobat_binned['Total Bio']
acrobat_binned['Percent Small'] = acrobat_binned['Small'] / acrobat_binned['Total Bio']
acrobat_binned['Percent Medium'] = acrobat_binned['Medium'] / acrobat_binned['Total Bio']
acrobat_binned['Percent Large'] = acrobat_binned['Large'] / acrobat_binned['Total Bio']
acrobat_binned.head()
```

```python
#Percentage of Size Groups over time
fig, ax = plt.subplots()
plt.scatter(acrobat_binned['Seconds from Start'],acrobat_binned['Percent Tiny'], c='b')
plt.scatter(acrobat_binned['Seconds from Start'],acrobat_binned['Percent Small'], c='r')
plt.scatter(acrobat_binned['Seconds from Start'],acrobat_binned['Percent Medium'], c='g')
plt.scatter(acrobat_binned['Seconds from Start'],acrobat_binned['Percent Large'], c='orange')
plt.xlabel('Seconds')
plt.ylabel('Percent of Total Biomass')
plt.legend();

#Percentage of Size Groups over Depth
fig, ax = plt.subplots()
plt.scatter(acrobat_binned['MiidasDepth (High)'],acrobat_binned['Percent Tiny'], c='b')
plt.scatter(acrobat_binned['MiidasDepth (High)'],acrobat_binned['Percent Small'], c='r')
plt.scatter(acrobat_binned['MiidasDepth (High)'],acrobat_binned['Percent Medium'], c='g')
plt.scatter(acrobat_binned['MiidasDepth (High)'],acrobat_binned['Percent Large'], c='orange')
plt.xlabel('Seconds')
plt.ylabel('Percent of Total Biomass')
plt.legend();
```
